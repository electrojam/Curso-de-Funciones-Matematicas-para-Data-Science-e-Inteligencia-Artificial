```python
import matplotlib.pyplot as plt
import numpy as np
```

##Función Lineal

f(x)=mx+b

```python
N = 100
m = -1
b = 3

x = np.linspace(-10,10, num=N)
```

```python
def f(x):
  return m*x+b

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```

##Funciones Polinómicas

P(x)=2x7−x4+3x2+4

Por ejemplo:

f(x)=x**2

El dominio de  f(x)=x**2  es  Domf=(−∞,∞) . Su imagen es  Imf=[0,∞)

```python
def f(x):
  return (2*x**7)-(x**4)+(3*x**2)+4

y = f(x)

plt.plot(x,y)
plt.grid()
```
