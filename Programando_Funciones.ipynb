```python
import matplotlib.pyplot as plt
import numpy as np
```

##Funci√≥n Lineal

f(x)=mx+b

```python
N = 100
m = -1
b = 3

x = np.linspace(-10,10, num=N)
```

```python
def f(x):
  return m*x+b

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```

##Funciones Polin√≥micas

P(x)=2x7‚àíx4+3x2+4

Por ejemplo:

f(x)=x**2

El dominio de  f(x)=x**2  es  Domf=(‚àí‚àû,‚àû) . Su imagen es  Imf=[0,‚àû)

```python
def f(x):
  return (2*x**7)-(x**4)+(3*x**2)+4

y = f(x)

plt.plot(x,y)
plt.grid()
```

### Funci√≥n logaritmo

El logaritmo est√° definido por la **relaci√≥n**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$

Logaritmo en base de x es igual a n.  O x es igual a b elevado a la n






```python
def f(x):
  return np.log2(x)

#Los valores de x deben ser diferentes de 0
#porque ninguna base elevada a la n dar√≠a 0
x = np.linspace(0.001,256, num=1000)

plt.plot(x,f(x))
```
