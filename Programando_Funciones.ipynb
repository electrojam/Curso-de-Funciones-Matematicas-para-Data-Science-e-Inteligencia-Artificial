### Función logaritmo

El logaritmo está definido por la **relación**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$

Logaritmo en base de x es igual a n.  O x es igual a b elevado a la n






```python
def f(x):
  return np.log2(x)

#Los valores de x deben ser diferentes de 0
#porque ninguna base elevada a la n daría 0
x = np.linspace(0.001,256, num=1000)

plt.plot(x,f(x))
```

# Función seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside:

$$H(x) =
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx, s in enumerate(x):
    if s>=0:
      Y[idx]=1
  return Y

N = 1000

x = np.linspace(-10,10, num=N)

y = H(x)

plt.plot(x,y)
```
