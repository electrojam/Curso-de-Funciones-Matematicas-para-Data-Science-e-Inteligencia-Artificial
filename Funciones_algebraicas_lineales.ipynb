```python
import matplotlib.pyplot as plt
import numpy as np
```

##Funci√≥n Lineal

f(x)=mx+b

```python
N = 100
m = -1
b = 3

def f(x):
  return m*x+b

x = np.linspace(-10,10, num=N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```
